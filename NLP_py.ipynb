```python
def scrape_article(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            # Replace 'article' with the appropriate HTML tag or class that contains the article content.
            article = soup.find('article')
            if article:
                return article.get_text()
        return "Article not found on this page"
    except Exception as e:
        return str(e)
```

```python
import pandas as pd
import re
import nltk
from nltk.tokenize import word_tokenize

# Download the 'punkt' tokenizer models
nltk.download('punkt')

# Define a regular expression pattern to remove special symbols
special_symbols_pattern = r'[^A-Za-z0-9\s]'

# Function to clean a single article
def clean_article(article):
    # Tokenization
    tokens = word_tokenize(article)

    # Remove special characters and convert to lowercase
    cleaned_tokens = [word.lower() for word in tokens if word.isalpha()]

    # Remove stop words
    cleaned_text = ' '.join([word for word in cleaned_tokens if word not in stop_words])

    # Remove newline characters
    cleaned_text = re.sub(r'\n', ' ', cleaned_text)

    # Remove special symbols
    cleaned_text = re.sub(special_symbols_pattern, '', cleaned_text)

    return cleaned_text

# Apply the cleaning function to the 'Scraped_Article' column in the DataFrame
df['Cleaned_Article'] = df['Scraped_Article'].apply(clean_article)

# Print the DataFrame with cleaned articles
print(df)
```

```python
# Read the Positive Words List from a file
with open("positive-words.txt", "r") as file:
    positive_words = set(word.strip() for word in file.readlines())

with open("negative-words.txt", "r", encoding="latin-1") as file:
    negative_words = set(word.strip() for word in file.readlines())
```

```python
# Create dictionaries to store positive and negative words
positive_word_dict = {}
negative_word_dict = {}

# Open and read the 'positive-words.txt' file with 'latin-1' encoding
with open("positive-words.txt", 'r', encoding='latin-1') as file:
    for line in file:
        word = line.strip()
        # Check if the word is not in the Stop Words List
        if word not in stop_words:
            positive_word_dict[word] = 1  # Assign a value (1) to indicate a positive word

# Open and read the 'negative-words.txt' file with 'latin-1' encoding
with open("negative-words.txt", 'r', encoding='latin-1') as file:
    for line in file:
        word = line.strip()
        # Check if the word is not in the Stop Words List
        if word not in stop_words:
            negative_word_dict[word] = -1  # Assign a value (-1) to indicate a negative word

# Print the positive and negative word dictionaries
print("Positive Words Dictionary:")
print(positive_word_dict)
print("\nNegative Words Dictionary:")
print(negative_word_dict)
```

```python
# Create a Master Dictionary
master_dictionary = {}

# Add positive words to the Master Dictionary
for word, value in positive_word_dict.items():
    master_dictionary[word] = value

# Add negative words to the Master Dictionary
for word, value in negative_word_dict.items():
    master_dictionary[word] = value
```

```python
# Function to calculate positive score for an article
def calculate_positive_score(article):
    # Tokenize the article
    tokens = word_tokenize(article)

    # Calculate positive score by summing up the values from the positive word dictionary
    positive_score = sum(master_dictionary.get(word, 0) for word in tokens)

    return positive_score

# Function to calculate negative score for an article
def calculate_negative_score(article):
    # Tokenize the article
    tokens = word_tokenize(article)

    # Calculate negative score by summing up the absolute values from the negative word dictionary
    negative_score = -sum(abs(master_dictionary.get(word, 0)) for word in tokens)

    return negative_score

# Apply the functions to the 'Cleaned_Article' column in the DataFrame
df['Positive_Score'] = df['Cleaned_Article'].apply(calculate_positive_score)
df['Negative_Score'] = df['Cleaned_Article'].apply(calculate_negative_score)

# Print the DataFrame with positive and negative scores
print(df[['Cleaned_Article', 'Positive_Score', 'Negative_Score']])
```

```python
# Function to calculate polarity score for an article
def calculate_polarity_score(positive_score, negative_score):
    # Calculate polarity score using the given formula
    polarity_score = (positive_score - negative_score) / ((positive_score + negative_score) + 0.000001)
    return polarity_score

# Function to calculate subjectivity score for an article
def calculate_subjectivity_score(positive_score, negative_score, total_words):
    # Calculate subjectivity score using the given formula
    subjectivity_score = (positive_score + negative_score) / ((total_words) + 0.000001)
    return subjectivity_score

# Apply the functions to the DataFrame
df['Polarity_Score'] = df.apply(lambda row: calculate_polarity_score(row['Positive_Score'], row['Negative_Score']), axis=1)
df['Subjectivity_Score'] = df.apply(lambda row: calculate_subjectivity_score(row['Positive_Score'], row['Negative_Score'], len(row['Cleaned_Article'].split())), axis=1)

# Print the DataFrame with polarity and subjectivity scores
print(df[['Cleaned_Article', 'Polarity_Score', 'Subjectivity_Score']])
```

```python
import nltk

# Download the 'averaged_perceptron_tagger' model for POS tagging
nltk.download('averaged_perceptron_tagger')

# Function to calculate average sentence length
def calculate_average_sentence_length(article):
    # Tokenize the article into sentences
    sentences = nltk.sent_tokenize(article)

    # Calculate the average sentence length
    total_words = sum(len(nltk.word_tokenize(sentence)) for sentence in sentences)
    total_sentences = len(sentences)

    if total_sentences == 0:
        return 0  # Avoid division by zero

    average_sentence_length = total_words / total_sentences
    return average_sentence_length

# Function to calculate percentage of complex words
def calculate_percentage_complex_words(article):
    # Tokenize the article into words
    words = nltk.word_tokenize(article)

    # Perform Part-of-Speech (POS) tagging
    pos_tags = nltk.pos_tag(words)

    # Count the number of complex words (words with tags starting with 'JJ' or 'RB')
    complex_word_count = sum(1 for _, pos_tag in pos_tags if pos_tag.startswith('JJ') or pos_tag.startswith('RB'))

    if len(words) == 0:
        return 0  # Avoid division by zero

    percentage_complex_words = complex_word_count / len(words)
    return percentage_complex_words

# Function to calculate Gunning Fox Index
def calculate_gunning_fox_index(average_sentence_length, percentage_complex_words):
    # Calculate Gunning Fox Index using the given formula
    gunning_fox_index = 0.4 * (average_sentence_length + percentage_complex_words)
    return gunning_fox_index

# Apply the functions to the DataFrame
df['Average_Sentence_Length'] = df['Cleaned_Article'].apply(calculate_average_sentence_length)
df['Percentage_Complex_Words'] = df['Cleaned_Article'].apply(calculate_percentage_complex_words)
df['Gunning_Fox_Index'] = df.apply(lambda row: calculate_gunning_fox_index(row['Average_Sentence_Length'], row['Percentage_Complex_Words']), axis=1)

# Print the DataFrame with readability analysis
print(df[['Cleaned_Article', 'Average_Sentence_Length', 'Percentage_Complex_Words', 'Gunning_Fox_Index']])
```

```python
# Function to calculate average number of words per sentence
def calculate_average_words_per_sentence(article):
    # Tokenize the article into sentences
    sentences = nltk.sent_tokenize(article)

    # Calculate the total number of words and sentences
    total_words = sum(len(nltk.word_tokenize(sentence)) for sentence in sentences)
    total_sentences = len(sentences)

    if total_sentences == 0:
        return 0  # Avoid division by zero

    average_words_per_sentence = total_words / total_sentences
    return average_words_per_sentence

# Apply the function to the DataFrame
df['Average_Words_Per_Sentence'] = df['Cleaned_Article'].apply(calculate_average_words_per_sentence)

# Print the DataFrame with the average number of words per sentence
print(df[['Cleaned_Article', 'Average_Words_Per_Sentence']])
```

```python
from nltk.corpus import cmudict

# Download the CMU Pronouncing Dictionary for syllable counting
nltk.download('cmudict')

# Function to count syllables in a word
def count_syllables(word):
    # Use the CMU Pronouncing Dictionary for syllable counting
    d = cmudict.dict()
    if word.lower() in d:
        return max([len(list(y for y in x if y[-1].isdigit())) for x in d[word.lower()]][0], 1)
    else:
        return 1  # For example, assume one syllable for unknown words

# Function to calculate the complex word count
def calculate_complex_word_count(article):
    # Tokenize the article into words
    words = nltk.word_tokenize(article)

    # Count the number of complex words (words with more than two syllables)
    complex_word_count = sum(1 for word in words if count_syllables(word) > 2)

    return complex_word_count

# Apply the function to the DataFrame
df['Complex_Word_Count'] = df['Cleaned_Article'].apply(calculate_complex_word_count)

# Print the DataFrame with the complex word count
print(df[['Cleaned_Article', 'Complex_Word_Count']])
```

```python
from nltk.corpus import stopwords
import string

# Download the stopwords dataset
nltk.download('stopwords')

# Function to calculate the word count
def calculate_word_count(cleaned_text):
    # Remove punctuation
    cleaned_text_no_punct = cleaned_text.translate(str.maketrans('', '', string.punctuation))

    # Tokenize the cleaned text into words
    words = nltk.word_tokenize(cleaned_text_no_punct)

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word.lower() not in stop_words]

    # Count the total words
    total_words = len(filtered_words)

    return total_words

# Apply the function to the DataFrame
df['Word_Count'] = df['Cleaned_Article'].apply(calculate_word_count)

# Print the DataFrame with the word count
print(df[['Cleaned_Article', 'Word_Count']])
```

```python
# Function to count syllables in each word
def count_syllables_per_word(word):
    # Handle exceptions for words ending with "es" and "ed"
    if word.endswith(("es", "ed")):
        return 0

    # Count vowels in the word
    vowels = "aeiouAEIOU"
    count = sum(1 for char in word if char in vowels)

    # Handle exceptions for words with no vowels
    if count == 0:
        return 1

    return count

# Function to calculate syllable count per word
def calculate_syllable_count_per_word(cleaned_text):
    # Tokenize the cleaned text into words
    words = nltk.word_tokenize(cleaned_text)

    # Count syllables for each word
    syllables_per_word = [count_syllables_per_word(word) for word in words]

    return syllables_per_word

# Apply the function to the DataFrame
df['Syllables_Per_Word'] = df['Cleaned_Article'].apply(calculate_syllable_count_per_word)

# Print the DataFrame with the syllable count per word
print(df[['Cleaned_Article', 'Syllables_Per_Word']])
```

```python
import re

# Function to calculate counts of personal pronouns
def calculate_personal_pronoun_counts(text):
    # Define a regex pattern for personal pronouns
    personal_pronoun_pattern = re.compile(r'\b(?:I|we|my|ours|us)\b', flags=re.IGNORECASE)

    # Find all matches of the pattern in the text
    matches = personal_pronoun_pattern.findall(text)

    # Count the occurrences of personal pronouns
    personal_pronoun_counts = len(matches)

    return personal_pronoun_counts

# Apply the function to the DataFrame
df['Personal_Pronoun_Count'] = df['Cleaned_Article'].apply(calculate_personal_pronoun_counts)

# Print the DataFrame with the counts of personal pronouns
print(df[['Cleaned_Article', 'Personal_Pronoun_Count']])
```

```python
# Function to calculate average word length
def calculate_average_word_length(cleaned_text):
    # Tokenize the cleaned text into words
    words = nltk.word_tokenize(cleaned_text)

    # Calculate the total number of characters in each word
    total_characters = sum(len(word) for word in words)

    # Calculate the total number of words
    total_words = len(words)

    if total_words == 0:
        return 0  # Avoid division by zero

    # Calculate average word length
    average_word_length = total_characters / total_words

    return average_word_length

# Apply the function to the DataFrame
df['Average_Word_Length'] = df['Cleaned_Article'].apply(calculate_average_word_length)

# Print the DataFrame with the average word length
print(df[['Cleaned_Article', 'Average_Word_Length']])
```
